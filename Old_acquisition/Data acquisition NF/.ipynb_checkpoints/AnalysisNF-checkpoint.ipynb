### Image visualization

```python
filt=2
```

```python
im=Get_image(runpath,point,filt)
dark=Get_dark(runpath,point,filt)
imd=Im_dark(runpath,point,filt)
f,ax=plt.subplots(1,2,figsize=(10,3))
ax[0].imshow(im)
ax[1].hist(im.flatten(),bins=100);
ax[1].set_yscale('log')
plt.suptitle('Image')

f,ax=plt.subplots(1,2,figsize=(10,3))
ax[0].imshow(dark)
ax[1].hist(dark.flatten(),bins=100);
ax[1].set_yscale('log')
plt.suptitle('Dark')

f,ax=plt.subplots(1,2,figsize=(10,3))
ax[0].imshow(imd)
ax[1].hist(imd.flatten(),bins=100);
ax[1].set_yscale('log')
plt.suptitle('Image-dark')

f,ax=plt.subplots(1,2,figsize=(10,3))
ax[0].imshow(mask*imd)
ax[1].hist((mask*imd).flatten(),bins=100);
print(np.sum(mask*imd))
ax[1].set_yscale('log')
plt.suptitle('Image-dark in ROI')
```
